###  gamma 

```python
import cv2
import numpy as np
import torch
from ultralytics import YOLO

# Check if PyTorch can see the GPU
print("CUDA Available:", torch.cuda.is_available())

# Load the YOLOv5 model with GPU support
model = YOLO('yolov5s.pt').to('cuda')  # Ensure the model is on GPU

# Precompute the gamma correction lookup table
def gamma_correction_table(gamma=1):
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255 for i in np.arange(0, 256)]).astype("uint8")
    return table

gamma_table = gamma_correction_table(1.4)

# Function to apply grayscale conversion, Gaussian smoothing, and Gamma transformation using LUT
def apply_enhancements(frame, gamma_table):
    # Convert to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Apply Gaussian smoothing
    smoothed = cv2.GaussianBlur(gray, (5, 5), 0)
    # Apply Gamma correction using the lookup table
    gamma_corrected = cv2.LUT(smoothed, gamma_table)
    # Convert back to BGR to keep the three channel format
    return cv2.cvtColor(gamma_corrected, cv2.COLOR_GRAY2BGR)

# Open the webcam
cap = cv2.VideoCapture('http://192.168.1.3:8080/video')
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

frame_skip = 3  # Adjust the number of frames to skip if necessary
frame_count = 0

while True:
    ret, frame = cap.read()
    if not ret:
        print("Error: Could not read frame.")
        break

    if frame_count % frame_skip == 0:
        # Convert frame to tensor and send to GPU
        frame_tensor = torch.from_numpy(frame).to('cuda')
        # Apply grayscale, Gaussian smoothing, and Gamma transformation using LUT
        processed_frame = apply_enhancements(frame_tensor.cpu().numpy(), gamma_table)

        # Perform object detection
        results = model(processed_frame)
        # Render detections on the frame
        annotated_frame = results[0].plot()

        # Display the frame
        cv2.imshow("Enhanced Night Vision with Gamma Correction and Object Detection", annotated_frame)

    frame_count += 1

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
 
```
